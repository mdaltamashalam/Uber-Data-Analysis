# **5. Preprocessing**
✅ Encode categorical features:

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Ensure df is your DataFrame, for example:
# df = pd.read_csv('your_data.csv')  # Replace with your actual DataFrame

label_encoders = {}

# Handle encoding for object-type columns
for col in df.select_dtypes(include='object').columns:
    try:
        # Drop NaN values temporarily for encoding
        df[col] = df[col].fillna('Missing')  # Optional: Replace NaN with 'Missing' or any other placeholder

        # Initialize LabelEncoder and fit/transform the column
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])

        # Store the LabelEncoder for possible inverse transformation later
        label_encoders[col] = le
    except Exception as e:
        print(f"Error encoding column {col}: {e}")

# Verify the changes by printing the first few rows
print(df.head())
```

✅ Scale numerical features:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load data (assuming 'df' is already loaded)

# Strip any leading or trailing spaces in column names
df.columns = df.columns.str.strip()

# Check column names to verify the correct column is present
print("Columns in DataFrame:", df.columns)

# Rename 'price' column to 'Price' for consistency
if 'price' in df.columns:
    df.rename(columns={'price': 'Price'}, inplace=True)

# Check if 'Price' exists after renaming
if 'Price' not in df.columns:
    print("Error: 'Price' column not found in the DataFrame!")
else:
    # Select only numeric columns for scaling
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Drop 'Price' column from X (features) and keep it in y (target)
    X = df[numeric_cols].drop('Price', axis=1, errors='ignore')  # Ensure 'Price' is not included in X
    y = df['Price']  # Target variable

    # Handle missing values if present (e.g., fill with mean)
    X.fillna(X.mean(), inplace=True)  # Fill NaN with mean for numeric columns

    # Initialize the StandardScaler
    scaler = StandardScaler()

    # Scale the features
    X_scaled = scaler.fit_transform(X)

    # Verify the scaling
    print(X_scaled[:5])  # Display the first 5 rows of scaled data

    # If needed, convert X_scaled back to a DataFrame (optional)
    X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

    # Check the first few rows
    print(X_scaled_df.head())
```

✅ Train-Test Split:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split  # Import train_test_split

# Load your data (assuming 'df' is already loaded)

# Strip any leading or trailing spaces in column names
df.columns = df.columns.str.strip()

# Check column names to verify the correct column is present
print("Columns in DataFrame:", df.columns)

# Rename 'price' column to 'Price' for consistency
if 'price' in df.columns:
    df.rename(columns={'price': 'Price'}, inplace=True)

# Check if 'Price' exists after renaming
if 'Price' not in df.columns:
    print("Error: 'Price' column not found in the DataFrame!")
else:
    # Select only numeric columns for scaling
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Drop 'Price' column from X (features) and keep it in y (target)
    X = df[numeric_cols].drop('Price', axis=1, errors='ignore')  # Ensure 'Price' is not included in X
    y = df['Price']  # Target variable

    # Handle missing values if present (e.g., fill with mean)
    X.fillna(X.mean(), inplace=True)  # Fill NaN with mean for numeric columns

    # Initialize the StandardScaler
    scaler = StandardScaler()

    # Scale the features
    X_scaled = scaler.fit_transform(X)

    # Verify the scaling
    print(X_scaled[:5])  # Display the first 5 rows of scaled data

    # If needed, convert X_scaled back to a DataFrame (optional)
    X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

    # Check the first few rows
    print(X_scaled_df.head())

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Verify the split
    print(f"Training data shape: X_train: {X_train.shape}, y_train: {y_train.shape}")
    print(f"Testing data shape: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

# **6. Model Building**
✅ Train different models:

```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# Train Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Train Random Forest model
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

# Train XGBoost model
xgb = XGBRegressor()
xgb.fit(X_train, y_train)

# Train LightGBM model
lgb = LGBMRegressor()
lgb.fit(X_train, y_train)

# Train CatBoost model
cat = CatBoostRegressor(verbose=0)
cat.fit(X_train, y_train)

# Predictions for testing
lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)
xgb_pred = xgb.predict(X_test)
lgb_pred = lgb.predict(X_test)
cat_pred = cat.predict(X_test)

# Optionally, calculate evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# For Linear Regression
print("Linear Regression:")
print(f"MAE: {mean_absolute_error(y_test, lr_pred)}")
print(f"MSE: {mean_squared_error(y_test, lr_pred)}")
print(f"R2: {r2_score(y_test, lr_pred)}")

# For Random Forest
print("\nRandom Forest:")
print(f"MAE: {mean_absolute_error(y_test, rf_pred)}")
print(f"MSE: {mean_squared_error(y_test, rf_pred)}")
print(f"R2: {r2_score(y_test, rf_pred)}")

# For XGBoost
print("\nXGBoost:")
print(f"MAE: {mean_absolute_error(y_test, xgb_pred)}")
print(f"MSE: {mean_squared_error(y_test, xgb_pred)}")
print(f"R2: {r2_score(y_test, xgb_pred)}")

# For LightGBM
print("\nLightGBM:")
print(f"MAE: {mean_absolute_error(y_test, lgb_pred)}")
print(f"MSE: {mean_squared_error(y_test, lgb_pred)}")
print(f"R2: {r2_score(y_test, lgb_pred)}")

# For CatBoost
print("\nCatBoost:")
print(f"MAE: {mean_absolute_error(y_test, cat_pred)}")
print(f"MSE: {mean_squared_error(y_test, cat_pred)}")
print(f"R2: {r2_score(y_test, cat_pred)}")
```
